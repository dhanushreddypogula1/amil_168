```python
#load the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("C:/Users/dhanu/Documents/data_clean.csv")
print(data)
```

```python
# Printing the information 
data.info()
```

```python
# Dataframe attributes
print(type(data))
print(data.shape)
print(data.size)
```

```python
# Drop dupplicate column( Temp C) and Unnamed column

# data1 = data.drop(['Unnamed: 0',"Temp C"], axis =1, inplace = True)
data1 = data.drop(['Unnamed: 0',"Temp C"], axis =1)
data1
```

```python
data1.info()
```

```python
# Convert the Month column data type to float data type

data1['Month']=pd.to_numeric(data['Month'],errors='coerce')
data1.info()
```

```python
# Print all duplicated rows
data1[data1.duplicated(keep = False)]
```

```python
# Checking for duplicated rows in the table
#Print only the duplicated row (one) only
data1[data1.duplicated()]
```

```python
# Drop duplicated rows
data1.drop_duplicates(keep='first', inplace = True)
data1
```

```python
# Change column names (Rename the columns)
data1.rename({'Solar.R': 'Solar'}, axis=1, inplace = True)
data1
```

#### Impute the missing values

```python
# Display data1 info()
data1.info()
```

```python
# Display data1 missing values count in each column using isnull().sum()
data1.isnull().sum()
```

```python
# Visualize data1 missing values 
cols = data1.columns
colours = ['black','yellow']
sns.heatmap(data1[cols].isnull(),cmap=sns.color_palette(colours),cbar=True)
```

```python
# Find the mean and median values of each numeric column
median_ozone = data1["Ozone"].median()
mean_ozone = data1["Ozone"].mean()
print("Median of Ozone:",median_ozone)
print("Mean of Ozone:", mean_ozone)
```

```python
# Find the mode values of categorical columns
data1['Ozone'] = data1['Ozone'].fillna(median_ozone)
data1.isnull().sum()
```

```python
# Impute missing values (Replace NaN with mean/median, mode etc.) using fillna()
median_solar = data1["Solar"].median()
mean_solar = data1["Solar"].mean()
print("Median of Solar:",median_solar)
print("Mean of Solar:",mean_solar)
```

```python
data1['Solar'] = data1['Solar'].fillna(median_ozone)
data1.isnull().sum()
```

```python
print(data1["Weather"].value_counts())
mode_weather = data1["Weather"].mode()[0]
print(mode_weather)
```

```python
data1["Weather"]=data1["Weather"].fillna(mode_weather)
data1.isnull().sum()
```

```python
print(data1["Month"].value_counts())
mode_month=data1["Month"].mode()[0]
print(mode_month)
```

```python
data1["Month"]=data1["Month"].fillna(mode_month)
data1.isnull().sum()
```

```python
data1.reset_index(drop=True)
```

```python
fig, axes = plt.subplots(2,1,figsize=(8,6),gridspec_kw={'height_ratios':[1,3]})
sns.boxplot(data=data1["Ozone"],ax=axes[0],color='skyblue',width=0.5,orient='h')
axes[0].set_title("Boxplot")
axes[0].set_xlabel("Ozone Levels")
sns.histplot(data1["Ozone"],kde=True,ax=axes[1],color='purple',bins=30)
axes[1].set_title("Histogram with KDE")
axes[1].set_xlabel("Ozone Levels")
axes[1].set_ylabel("Frequency")
plt.tight_layout()
plt.show()
```
